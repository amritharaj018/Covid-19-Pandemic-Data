# BAR CHART

```python
fig,ax=plt.subplots(1,3,figsize = (25,15))
bar_height = 0.35
bar_width = 1.0

ax[0].bar(data1["Country"],data1["Population"],color="cyan", edgecolor ="black")
ax[0].set_xlabel("Country",fontsize=22,fontweight="bold")
ax[0].set_ylabel("Population",fontsize=22,fontweight="bold")
ax[0].set_title("Population v/s Country",fontsize=17,fontweight="bold")
ax[0].set_xticks(data1["Country"])

ax[0].set_xticklabels(data1["Country"],rotation=75,fontsize=15)


ax[1].bar(data1["Country"],data1["Total Cases"],color="yellow", edgecolor ="black")
ax[1].set_xlabel("Country",fontsize=22,fontweight="bold")
ax[1].set_ylabel("Total Cases",fontsize=22,fontweight="bold")
ax[1].set_title("Total Cases v/s Country",fontsize=17,fontweight="bold")
ax[1].set_xticks(data1["Country"])

ax[1].set_xticklabels(data1["Country"],rotation=75,fontsize=15)


ax[2].bar(data1["Country"],data1["Death percentage"],color="red", edgecolor ="black")
ax[2].set_xlabel("Country",fontsize=22,fontweight="bold")
ax[2].set_ylabel("Death percentage",fontsize=22,fontweight="bold")
ax[2].set_title("Death Percentage v/s Country",fontsize=17,fontweight="bold")
ax[2].set_xticks(data1["Country"])

ax[2].set_xticklabels(data1["Country"],rotation=75,fontsize=15)

plt.subplots_adjust(hspace =0.5)
plt.tight_layout()
plt.show()
```

# SCATTER PLOT

```python
fig, ax = plt.subplots(figsize=(10,5),sharey=True)
ds = data.iloc[150:180]
ax.scatter(ds["Country"],ds["Population"],color="red", label="Population", alpha=1 , s= 80)
ax.scatter(ds["Country"],ds["Total Deaths"],color="green", label="Total_Deaths", alpha=1, s= 80,zorder=2)
ax.scatter(ds["Country"],ds["Total Cases"],color="orange", label="Total Cases", alpha=1, s= 80,zorder=1)


ax.set_xlabel("Country",fontweight="bold")
ax.set_xticks(ds["Country"])
ax.set_ylabel("Frequency",fontweight="bold")
ax.legend()
plt.xticks(rotation=90)
ax.set_title("Population/Total Deaths/Total Cases v/s Country",fontsize=17,fontweight="bold")
ax.set_xticklabels(ds["Country"], rotation=90)
plt.show()
```

# HEAT MAP

```python
import seaborn as sns
```

```python
s_cols=["Population","Total Deaths","Total Cases"]
s_data= data[s_cols]
correlation_matrix= s_data.corr(method="spearman")
plt.title("Correlation Matrix")
print(correlation_matrix)
sns.heatmap(correlation_matrix, cmap="PiYG",annot=True)
plt.title("Heatmap Showing Correlation between Population, Total Deaths & Total Cases")
plt.show()
```
